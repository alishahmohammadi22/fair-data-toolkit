## 1. The FAIR Principles — Origin and Context

The **FAIR Data Principles** were published in 2016 by Mark Wilkinson and colleagues in  
*Scientific Data* (Nature): [doi:10.1038/sdata.2016.18](https://doi.org/10.1038/sdata.2016.18)

FAIR stands for:

| Letter | Principle | Core idea |
|--------|-----------|-----------|
| **F** | **Findable** | Data and metadata have persistent, globally unique identifiers and are indexed in searchable resources |
| **A** | **Accessible** | Data and metadata are retrievable via standardised, open protocols — even if the data itself is restricted |
| **I** | **Interoperable** | Data uses formal, shared knowledge representations and FAIR vocabularies |
| **R** | **Reusable** | Data has rich metadata, clear licences, and meets community standards — enabling legitimate reuse |

The original paper listed **15 sub-principles** spanning both data and metadata.  
The principles were designed to be **machine-actionable** — not just readable by humans.

````python
# Let's explore the FAIR toolkit and the RDA indicators
# pip install -e ".[notebooks]" from the fair-data-toolkit repo root

from fair_toolkit import RDA_INDICATORS, FAIRPrinciple, get_indicators_by_principle
from collections import Counter

# How many indicators per FAIR principle?
counts = Counter(ind.principle.value for ind in RDA_INDICATORS)
print("RDA FAIR Maturity Model — indicator counts by principle:")
for principle in ["F", "A", "I", "R"]:
    print(f"  {principle}: {counts[principle]} indicators")
print(f"  Total: {sum(counts.values())} indicators")
````

## 2. FAIR ≠ Open

This is one of the most important distinctions for pharmaceutical organisations:

**FAIR does NOT require open access.**

The Accessible principle (A) is about having a *defined, standardised access mechanism* —  
not about making data freely downloadable. A dataset behind a Data Access Committee (DAC)  
or a Material Transfer Agreement (MTA) can be fully FAIR if:

- Its metadata is publicly findable and accessible (the metadata is open, the data is not)
- The access protocol is documented and standardised (OAuth, DAC request form)
- The licence clearly states the conditions under which access is granted

```
FAIR spectrum:

  Not FAIR ──────────────────────────────────── Fully FAIR
     ↑                                              ↑
  Silo'd data,                              Machine-actionable,
  no metadata,                              rich metadata, PIDs,
  no identifiers                            standard formats,
                                            clear governance

  Open spectrum (independent axis):

  Fully Restricted ──────────────────────── Fully Open
     ↑                                          ↑
  Trade secrets,                           CC0 / CC BY,
  patient data,                            pre-competitive data,
  safety databases                         published datasets
```

A dataset can be:
- **FAIR and Open** — ideal for pre-competitive data (e.g., ChEMBL, PDB)
- **FAIR and Restricted** — clinical trials, safety databases, proprietary assays
- **Open but not FAIR** — a ZIP file on a web server with no metadata
- **Neither** — data locked in a vendor LIMS with no external access or documentation

````python
# Visualise the FAIR spectrum
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

fig, ax = plt.subplots(figsize=(9, 6))
ax.set_xlim(0, 10); ax.set_ylim(0, 10)
ax.set_xlabel("Open (0 = fully restricted, 10 = fully open)", fontsize=12)
ax.set_ylabel("FAIR (0 = not FAIR, 10 = fully FAIR)", fontsize=12)
ax.set_title("The FAIR × Open Space\n(Any quadrant can be valid — FAIRness is independent of Openness)", fontsize=12)
ax.axhline(5, color='gray', linestyle='--', alpha=0.4)
ax.axvline(5, color='gray', linestyle='--', alpha=0.4)

examples = [
    (1, 1, "Internal LIMS\n(no metadata, no API)", "red"),
    (9, 1, "Public ZIP on web server\n(open but no metadata)", "orange"),
    (2, 9, "Clinical trial dataset\n(controlled access, fully FAIR)", "blue"),
    (9, 9, "ChEMBL / PDB\n(open + fully FAIR)", "green"),
    (5, 5, "Most pharma\ndata today", "gray"),
]

for x, y, label, colour in examples:
    ax.scatter(x, y, s=120, color=colour, zorder=5)
    ax.annotate(label, (x, y), textcoords="offset points", xytext=(8, 5), fontsize=9)

ax.set_facecolor("#f8f9fa")
plt.tight_layout()
plt.savefig("../docs/fair_open_quadrant.png", dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved.")
````

## 3. The 15 Sub-Principles

The original Wilkinson et al. paper defined 15 sub-principles organised under F, A, I, R.
The RDA FAIR Maturity Model later expanded these into 41 measurable indicators  
(covered in detail in Article 02).

### Findable (F1–F4)
- **F1** — (Meta)data are assigned a globally unique and persistent identifier
- **F2** — Data are described with rich metadata
- **F3** — Metadata clearly and explicitly includes the identifier of the data it describes
- **F4** — (Meta)data are registered or indexed in a searchable resource

### Accessible (A1–A2)
- **A1** — (Meta)data are retrievable by their identifier using a standardised communication protocol
  - **A1.1** — The protocol is open, free, and universally implementable
  - **A1.2** — The protocol allows for authentication and authorisation, where necessary
- **A2** — Metadata should be accessible even when the data is no longer available

### Interoperable (I1–I3)
- **I1** — (Meta)data use a formal, accessible, shared, and broadly applicable language for knowledge representation
- **I2** — (Meta)data use vocabularies that follow FAIR principles
- **I3** — (Meta)data include qualified references to other (meta)data

### Reusable (R1–R1.3)
- **R1** — (Meta)data are richly described with a plurality of accurate and relevant attributes
  - **R1.1** — (Meta)data are released with a clear and accessible data usage licence
  - **R1.2** — (Meta)data are associated with detailed provenance
  - **R1.3** — (Meta)data meet domain-relevant community standards

````python
# Show the sub-principle breakdown from the RDA Maturity Model
from fair_toolkit import RDA_INDICATORS, EvidenceLevel
from collections import defaultdict

# Group indicators by sub-principle
by_sub = defaultdict(list)
for ind in RDA_INDICATORS:
    by_sub[ind.sub_principle.value].append(ind)

print(f"{'Sub-principle':<12} {'Count':>6}  {'Essential':>9}  {'Important':>9}  {'Useful':>6}")
print("-" * 52)
for sp, inds in sorted(by_sub.items()):
    essential = sum(1 for i in inds if i.priority == EvidenceLevel.ESSENTIAL)
    important = sum(1 for i in inds if i.priority == EvidenceLevel.IMPORTANT)
    useful    = sum(1 for i in inds if i.priority == EvidenceLevel.USEFUL)
    print(f"{sp:<12} {len(inds):>6}  {essential:>9}  {important:>9}  {useful:>6}")
````

## 4. Why FAIR Matters in Pharmaceutical R&D

The pharmaceutical industry generates enormous volumes of heterogeneous data:  
genomics, proteomics, metabolomics, assay data, clinical trials, imaging, manufacturing.  

The cost of **not** being FAIR is high:

### 4.1 Internal Costs
- **Data silos**: 70–80% of research data in pharma is never reused after the study closes
- **Reproducibility failures**: lack of metadata makes it impossible to reproduce results years later  
- **Regulatory risk**: FDA, EMA, and ICH guidelines increasingly require FAIR-compliant data for submissions  
- **Decision latency**: finding the right dataset for a compound decision takes weeks instead of minutes

### 4.2 Industry-Wide Costs
- **Duplicated experiments**: the same assay is run multiple times across companies because prior data is inaccessible  
- **Slow drug discovery**: pre-competitive data sharing initiatives (TransCelerate, ATOM, IMI) require FAIR data as a prerequisite  
- **Patient risk**: incomplete safety data from non-FAIR databases may miss adverse event signals

### 4.3 The Business Case for FAIR
Estimated return on FAIR investment for a large pharmaceutical organisation:

| Benefit | Estimated Value |
|---------|-----------------|
| Reduced data re-generation | €10–50M per year |
| Faster regulatory submissions | 6–12 months cycle time reduction |
| AI/ML readiness (FAIR data as training data) | Enables next-generation drug discovery platforms |
| Cross-company pre-competitive data sharing | Access to multi-billion compound datasets |

## 5. Key FAIR Stakeholders and Resources

### GO FAIR International Support and Coordination
- Website: [go-fair.org](https://www.go-fair.org)
- Maintains the authoritative FAIR principles definition
- Publishes implementation profiles and maturity indicators

### Research Data Alliance (RDA)
- Website: [rd-alliance.org](https://www.rd-alliance.org)
- Published the **RDA FAIR Data Maturity Model** (2020) — the focus of Article 02
- 41 measurable indicators with priority levels (Essential / Important / Useful)

### Pistoia Alliance
- Website: [pistoiaalliance.org](https://www.pistoiaalliance.org)
- Published the **FAIR Maturity Matrix** specifically for pharmaceutical organisations
- 4-level tiered model (Bronze/Silver/Gold/Platinum) — focus of Article 03

### FAIRsharing
- Website: [fairsharing.org](https://fairsharing.org)
- Registry of standards, databases, and policies relevant to FAIR implementation
- Essential for finding the right community standard for your data type

### FAIR Maturity Indicators (FAIR Metrics Group)
- Published by Wilkinson, Dumontier et al. (2019)
- Provides machine-actionable metrics for automated FAIR assessment
- Foundation of the agentic FAIR scorer (covered in future articles)
