## 1. Why a Pharmaceutical-Specific Framework?

The RDA FAIR Maturity Model is designed to be domain-agnostic.  
While powerful, it does not address pharma-specific concerns like:

- **GxP compliance** (21 CFR Part 11, EU Annex 11, ICH Q10)
- **CDISC standards** (SDTM, ADaM) for clinical data
- **Pre-competitive data sharing** (TransCelerate, ATOM, IMI)
- **Regulatory submission requirements** (FDA, EMA eCTD)
- **Proprietary instrument data** (Waters, Agilent, Thermo formats)
- **Material Transfer Agreements** and data access governance

The **Pistoia Alliance FAIR Maturity Matrix** addresses these gaps by providing  
a tiered model built specifically for pharmaceutical and life sciences organisations.

### The Four Maturity Levels

| Level | Name | Core Characteristic |
|-------|------|---------------------|
| 🥉 **Bronze** | Awareness & Initial Implementation | Data has internal IDs, basic metadata captured, documented access process exists |
| 🥈 **Silver** | Defined Processes & Standards | Global PIDs, domain community standards adopted, APIs documented, standard licences |
| 🥇 **Gold** | Automated Assessment & Compliance | Machine-readable metadata, automated validation, rich provenance, FAIR scores tracked |
| 💎 **Platinum** | Self-Service FAIR & Continuous Improvement | FAIR-by-design, machine-actionable governance, cross-org interoperability, automated scoring |

```python
# Explore the Pistoia Matrix
from fair_toolkit import PISTOIA_MATRIX, MaturityLevel, describe_level
import pandas as pd

# Summary of indicators per level and dimension
records = []
for ind in PISTOIA_MATRIX:
    records.append({
        "ID": ind.id,
        "Level": ind.level.value,
        "Dimension": ind.dimension,
        "Name": ind.name,
        "RDA Links": len(ind.rda_indicator_ids),
    })

df = pd.DataFrame(records)
print("Pistoia FAIR Maturity Matrix — Indicator Summary")
print()
print(df.groupby(["Level", "Dimension"])["ID"].count().unstack(fill_value=0))
print()
print(f"Total Pistoia indicators: {len(df)}")
```

## 2. Level Descriptions

### 🥉 Bronze — Awareness & Initial Implementation

At Bronze level, the organisation:
- Assigns **internal identifiers** to all datasets
- Captures a **minimum metadata set** (title, creator, date, data type, project)
- Has a **documented data access process** (even if informal)
- Stores data in **non-proprietary formats** (or provides an open-format export)
- Has made a **licensing decision** for all datasets (even if restricted)

**Where most pharma organisations are today** — especially for non-GxP research data.

---

### 🥈 Silver — Defined Processes & Standards

At Silver level:
- **Global PIDs** (DOIs) assigned to externally shareable datasets
- **Controlled vocabulary mapping** — key fields use OBI, ChEBI, NCBITaxon, etc.
- **Documented, stable API** for programmatic access to key data systems
- **Standard licences** (CC BY, CC0) for open datasets
- **Provenance captured** (who/what/when) for all data transformations
- **Community standard compliance** validated for at least one primary data type

**Where leading pharma data organisations are targeting** as a 2–3 year goal.

---

### 🥇 Gold — Automated Assessment & Compliance

At Gold level:
- **Machine-readable metadata** validated automatically at submission
- **FAIR scores calculated** and tracked for all new datasets
- **Rich cross-references** in metadata and data (protocols.io, ChEMBL, UniProt)
- **Metadata tombstones** when data is retired
- **Automated compliance validation** against community standards
- **Provenance in PROV-O** or domain-specific standard (ISA-Tab, BIDS)

**Where top-tier pharma data platforms aspire to be** — typically 5+ years away for most.

---

### 💎 Platinum — Self-Service FAIR & Continuous Improvement

At Platinum level:
- **FAIR-by-design**: metadata auto-generated at data capture
- **Machine-actionable governance**: access control via policy-as-code (OPA)
- **Cross-organisational interoperability**: data flows to external partners without manual transformation
- **Continuous improvement loops** driven by automated FAIR scoring
- **Contributing to industry standards** development

**The aspirational target** — achieved by best-in-class repositories (ChEMBL, PDB, EBI).

```python
# Simulate a maturity assessment for a pharma organisation
from fair_toolkit import PISTOIA_MATRIX, MaturityLevel
from fair_toolkit.models.pistoia_matrix import get_matrix_by_level

# Define which Pistoia indicators an organisation currently meets
# (True = criterion met, False = not yet met)
current_state = {
    # Bronze
    "PA-F-B1": True,   # Internal IDs: YES — all datasets have LIMS IDs
    "PA-F-B2": True,   # Basic metadata: YES — ELN captures title, creator, date
    "PA-A-B1": True,   # Access process: YES — "email the data owner" process documented
    "PA-I-B1": True,   # Non-proprietary format: YES — CSV export available
    "PA-R-B1": True,   # Licence decision: PARTIALLY — most datasets have "internal use" decision

    # Silver
    "PA-F-S1": False,  # Global PIDs: NO — no DOIs assigned yet
    "PA-F-S2": False,  # External index: NO — not in any external catalogue
    "PA-A-S1": True,   # Documented API: YES — LIMS REST API documented internally
    "PA-A-S2": True,   # Standards-based access control: YES — Azure AD SSO
    "PA-I-S1": False,  # Controlled vocabularies: NO — using internal codebooks
    "PA-I-S2": False,  # Machine-readable metadata: NO — metadata only in LIMS UI
    "PA-R-S1": False,  # Standard licences: NO — custom IP agreements only
    "PA-R-S2": True,   # Provenance captured: YES — ELN audit trail (ALCOA+)

    # Gold
    "PA-F-G1": False,
    "PA-A-G1": False,
    "PA-I-G1": False,
    "PA-R-G1": False,
    "PA-R-G2": False,

    # Platinum
    "PA-F-P1": False,
    "PA-A-P1": False,
    "PA-I-P1": False,
    "PA-R-P1": False,
}

# Calculate level achievement
for level in [MaturityLevel.BRONZE, MaturityLevel.SILVER, MaturityLevel.GOLD, MaturityLevel.PLATINUM]:
    level_inds = get_matrix_by_level(level)
    met = sum(1 for i in level_inds if current_state.get(i.id, False))
    total = len(level_inds)
    pct = round(met / total * 100)
    bar = "█" * (pct // 5) + "░" * (20 - pct // 5)
    icon = {"bronze": "🥉", "silver": "🥈", "gold": "🥇", "platinum": "💎"}[level.value]
    print(f"  {icon} {level.value.upper():8s}: [{bar}] {pct:3d}%  ({met}/{total} criteria met)")

print()
print("Current maturity level: 🥉 Bronze (completing Silver)")
print("Recommended next steps:")
print("  1. Assign DOIs to all externally shareable datasets (PA-F-S1)")
print("  2. Map key metadata fields to OBI/ChEBI/NCBITaxon (PA-I-S1)")
print("  3. Apply standard open licences to pre-competitive datasets (PA-R-S1)")
print("  4. Register datasets in an external catalogue (PA-F-S2)")
```

## 3. Mapping Pistoia Levels to RDA Indicators

The Pistoia Matrix and the RDA FAIR Maturity Model are complementary.  
The Pistoia Matrix provides a **tiered progression framework** while the  
RDA model provides the **detailed measurement criteria**.

### How to use both together:

1. **Start with Pistoia** to determine your current maturity level and set a target
2. **Use the RDA model** to drill down into exactly which indicators need to improve
3. **Track RDA scores** over time as you progress through Pistoia levels

### Level → RDA Indicator Mapping (key examples)

| Pistoia Level | Pistoia Criterion | Key RDA Indicators |
|---------------|-------------------|---------------------|
| Bronze F-B1 | Internal identifiers | F1-01D (PID on data) |
| Bronze F-B2 | Basic metadata | F2-01M (rich metadata) |
| Silver F-S1 | Global PIDs | F1-01D, F1-02D, F1-01M, F1-02M |
| Silver I-S1 | Controlled vocabularies | I2-01M, I2-01D |
| Gold R-G2 | Community standard compliance | R1.3-01D, R1.3-02D |
| Platinum R-P1 | Automated FAIR scoring | All 41 indicators |
