The **Pistoia Alliance FAIR Maturity Matrix** is an *organisational* maturity model
for FAIR implementation in life science organisations.

> **Key concept:** Unlike the RDA FAIR Maturity Model (Article 2), which evaluates
> *individual datasets*, the Pistoia Alliance model evaluates an *organisation's*
> overall FAIR capability. It answers: _"How mature is your organisation's ability
> to create and maintain FAIR data?"_

### Framework at a glance

| | |
|---|---|
| **Version** | 1.1, 2025-03-14 |
| **Licence** | CC BY 4.0 |

| **Created by** | 20+ experts from AstraZeneca, BMS, Novo Nordisk, ELIXIR, GO FAIR Foundation, ORCID, and others |
| **Structure** | **7 dimensions** (rows) × **6 maturity levels** (columns) |

The matrix is intentionally *descriptive, not prescriptive* — it provides a framework
for self-assessment and shared conversation, not a pass/fail checklist.

## The Matrix Structure

### 6 Maturity Levels (columns)

| Level | Nickname | Marketplace Metaphor | Key Features |
|-------|----------|---------------------|--------------|
| **L0** | Life is unFAIR | Junkyard | Lack of awareness, possibly acquiring awareness |
| **L1** | Started the FAIR journey | Flea market | Awareness started; first pilots for implementation |
| **L2** | Getting FAIR | Street Market | Pilots for implementation are in place |
| **L3** | Pretty FAIR | Specialized Local Markets | Transition to good and best practice |
| **L4** | Really FAIR | Hyper Market | Operational, best practice known. Internal org focus. Emerging cross-org. |
| **L5** | FAIRest of them all | Digital Online Store | Aspirational. Cross-org standards and interoperability. |

> **Levels are cumulative:** Level N typically encompasses capabilities from Level N-1.
> The "marketplace" metaphors were introduced by AstraZeneca team members.

### 7 Dimensions (rows)

| Dimension | Key Notions |
|-----------|-------------|
| **FAIR data** | Data, metadata, and data products |
| **FAIR leadership** | Types of leadership necessary for FAIR implementation |
| **FAIR strategy** | Now, the next FAIR stage, and how to get there |
| **FAIR roles** | What kind of (human) roles are necessary to implement FAIR |
| **FAIR processes** | Which processes must we explicitly implement |
| **FAIR knowledge** | What needs to be known for FAIR implementation |
| **FAIR tools and infrastructures** | From persistent identifiers to controlled vocabularies to semantic models |

## Organisational Self-Assessment

You can use the matrix as a self-assessment tool by scoring your organisation
against each of the 7 dimensions. The result shows your maturity profile —
you may be at different levels in different dimensions, which is expected and normal.

```python
# Simple self-assessment: score your organisation per dimension
from fair_toolkit.models.pistoia_matrix import MaturityLevel, MatrixDimension, get_cell

# Example scores (L0–L5) — edit these to reflect your organisation
my_scores = {
    MatrixDimension.DATA:       MaturityLevel.L2,
    MatrixDimension.LEADERSHIP: MaturityLevel.L1,
    MatrixDimension.STRATEGY:   MaturityLevel.L2,
    MatrixDimension.ROLES:      MaturityLevel.L1,
    MatrixDimension.PROCESSES:  MaturityLevel.L2,
    MatrixDimension.KNOWLEDGE:  MaturityLevel.L1,
    MatrixDimension.TOOLS:      MaturityLevel.L2,
}

print("ORGANISATIONAL FAIR MATURITY PROFILE")
print("=" * 70)
print(f"{'Dimension':<35} {'Level':<6} {'Nickname'}")
print("-" * 70)

level_nums = {l: i for i, l in enumerate(MaturityLevel)}
total = 0
for dim, level in my_scores.items():
    meta = LEVEL_METADATA[level]
    num = level_nums[level]
    bar = "█" * (num + 1) + "░" * (5 - num)
    total += num
    print(f"{dim.value:<35} {level.value:<6} {bar}  {meta['nickname']}")

avg = total / len(my_scores)
avg_level = list(MaturityLevel)[round(avg)]
print("-" * 70)
print(f"{'Average':35} {avg:.1f}/5  → nearest: {avg_level.value} \"{LEVEL_METADATA[avg_level]['nickname']}\"")
print()
print("NOTE: Having different levels per dimension is normal. Focus improvement")
print("efforts on the lowest-scoring dimensions that are most strategically important.")
```

## The FAIR Tools & Infrastructures Dimension

This dimension has a unique property: it includes sub-indicators for each
FAIR letter (F/A/I/R), showing how tools and infrastructure contribute to each
FAIR principle. This bridges the organisational maturity model with the data-level
FAIR principles.

```python
# The FAIR Tools dimension has F/A/I/R sub-indicators
# Here we show how tools evolve across levels

tools_progression = {
    "L0": {"F": "None", "A": "None", "I": "None", "R": "None"},
    "L1": {
        "F": "URI/GUPRI policy; metadata catalogs; identifier policy for key datasets",
        "A": "Understanding of data challenges; gap assessment for tooling",
        "I": "Interoperability maturity indicators tracked",
        "R": "Reusability maturity indicators tracked",
    },
    "L2": {
        "F": "Identifier services; metadata schemas; registries",
        "A": "Auth & authorisation services; access controls",
        "I": "Less in focus at this stage",
        "R": "Less in focus at this stage",
    },
    "L3": {
        "F": "Domain-specific identifier registries; FIP tools",
        "A": "Controlled access systems; data governance platforms",
        "I": "Domain data models; ontology tools; CDISC/ISA-Tab converters",
        "R": "Licence management tools; provenance tracking",
    },
    "L4": {
        "F": "Enterprise-level registries; GUPRI consistently implemented",
        "A": "Federated + centralised access; automated licence checks",
        "I": "Knowledge graphs; semantic web technologies",
        "R": "Automated FAIRification tools; reuse metrics",
    },
    "L5": {
        "F": "Cross-org identifier standards; self-describing objects",
        "A": "Ecosystem-level access standards",
        "I": "Machine-enabled AI and semantic solutions",
        "R": "Automated provenance; cross-org reuse at scale",
    },
}

print("FAIR Tools & Infrastructures — F/A/I/R Sub-indicators by Level")
print("=" * 80)
for level_str, subs in tools_progression.items():
    meta = LEVEL_METADATA[MaturityLevel(level_str)]
    print(f"\n{level_str}: \"{meta['nickname']}\"")
    for principle, desc in subs.items():
        print(f"  [{principle}] {desc}")
```

## Connection to the RDA FAIR Maturity Model

The Pistoia Alliance and RDA frameworks are complementary:

| Aspect | RDA FAIR Maturity Model | Pistoia Alliance FMM |
|--------|------------------------|----------------------|
| **Focus** | Individual datasets | Organisation as a whole |
| **Structure** | 41 indicators across F/A/I/R | 7 dimensions × 6 levels |
| **Use** | Dataset-level compliance scoring | Organisational capability assessment |
| **Output** | Compliance score per indicator | Maturity level per dimension |
| **Context** | Any research dataset | Life science organisations |

A mature organisation (L4–L5 Pistoia) will typically also produce datasets that
score highly on RDA indicators — but the relationship is not 1-to-1. An organisation
can produce FAIR-compliant datasets in some areas while being at L1–L2 overall.

### A Practical Approach

1. Use **Pistoia Alliance FMM** to assess your organisation's *capability* and plan investments
2. Use **RDA FAIR Maturity Model** to assess specific datasets and prioritise FAIRification
3. Both frameworks should inform each other in a continuous improvement cycle

```python
# Show corresponding Pistoia levels and what RDA compliance you'd expect
print("Approximate relationship between Pistoia levels and RDA compliance:")
print()

roadmap = {
    "L0": "Very few RDA indicators met. Data is not systematically described or findable.",
    "L1": "F-indicators partially met (identifiers starting). A/I/R largely not met.",
    "L2": "F-indicators largely met. A-indicators emerging. I/R indicators limited.",
    "L3": "F+A indicators well-covered. I-indicators emerging locally. R partial.",
    "L4": "F+A+I indicators largely met. R-indicators strong. GUPRIs consistent.",
    "L5": "All F/A/I/R indicators met. Machine-actionable. Ecosystem interoperability.",
}

for level_str, rda_map in roadmap.items():
    meta = LEVEL_METADATA[MaturityLevel(level_str)]
    print(f"{level_str} \"{meta['nickname']}\":")
    print(f"  {rda_map}")
    print()

print("Source: Pistoia Alliance FAIR Maturity Matrix v1.1")
print("        https://pistoiaalliance.github.io/FAIRMaturityMatrix/")
```
